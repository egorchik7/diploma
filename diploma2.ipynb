```python
from scipy.optimize import newton

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Определение функции уравнения
def equation(x, s):
    return np.tan(x) - (np.tanh(s) / s) * x

# Массив значений a
s_values = np.arange(0, 10, 0.01)

# Список для хранения корней
roots = []

# Cписок для хранения выражения s^2 + x^2
lambds = []

# Цикл по параметру s
for s in s_values:
    initial_guess = 4.4  # начальное приближение (может быть подобрано опытным путем)
    try:
        # Поиск корня с использованием метода Ньютона
        root = newton(equation, initial_guess, args=(s,))
        roots.append(root)
        lambds.append(root**2 + s**2)
    except RuntimeError as e:
        # В случае неудачи добавим NaN в список
        roots.append(np.nan)
        lambds.append(np.nan)
        print(f"Не удалось найти корень при s={s}: {e}")

# Создание графика
plt.figure(figsize=(10, 5))
plt.plot(s_values, roots, marker='o',markersize=1, linestyle='-', linewidth=1)
plt.title(r"Первый положительный корень уравнения $tg \Lambda = \frac {th s}{s} \Lambda $ в зависимости от параметра s")
plt.xlabel("Волновое число s")
plt.ylabel("$\Lambda$")
plt.axhline(y=3.14, color='red', linewidth=1)
plt.text(0, 3.14, '$\Lambda = \pi$', verticalalignment='bottom', color='red')
plt.axhline(y=4.493, color='red', linewidth=1)
plt.text(0, 4.493, '$\Lambda = 4.493$', verticalalignment='bottom', color='red')

plt.grid(True)
#plt.savefig('lambda_from_s.png')
plt.show()

# Создание 2 графика
plt.figure(figsize=(10, 5))
plt.plot(s_values, lambds, marker='o',markersize=1, linestyle='-', linewidth=1)
plt.title("$ \Lambda^{2} + s^{2}$ в зависимости от параметра s")
plt.xlabel("Волновое число s")
plt.ylabel("$\Lambda^2 + s^2$")
plt.axhline(y=20.19, color='red', linewidth=1)
plt.text(7, 20.19, '$\Lambda^{2} + s^2 = 20.19$', verticalalignment='bottom', color='red')
plt.grid(True)
plt.savefig('lambda^2_s^2_from_s.png')
plt.show()
```
